```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('pose_estimation.csv')
```

```python
df['vx'].plot()
```

```python
vx = np.cumsum(df['ax']) * (1 / 10)  # Assuming a time step of 0.01 seconds (1/100 Hz)
vy = np.cumsum(df['ay']) * (1 / 10)
```

```python
x = np.cumsum(df['vx']) * (1 / 10)  # Assuming a time step of 0.01 seconds (1/100 Hz)
y = np.cumsum(df['vy']) * (1 / 10)
```

```python
plt.plot(x, y)

plt.axis('equal')
```
